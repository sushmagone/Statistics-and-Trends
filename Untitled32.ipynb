```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read data from csv
data = pd.read_csv('Restaurant_revenue .csv')
data 
```

```python
# drop duplicates
data.drop_duplicates(inplace=True)
```

```python
# 1. Inspect for missing values
print("Missing Values before Cleaning:")
print(data.isnull().sum())
```

```python
print("\nData after Cleaning:")
print("Missing Values after Cleaning:")
print(data.isnull().sum())
print("Data Types after Cleaning:")
print(data.dtypes)
print("Number of Duplicates after Cleaning:", data.duplicated().sum())
```

```python
def describe_statistics(data):
    """Displays the descriptive statistics for the dataset."""
    return data.describe()
describe_statistics(data)
```

```python
def correlation_statistics(data):
    """Displays the correlation matrix for numerical features."""
    return data.corr(numeric_only=True)
correlation_statistics(data)
```

```python
numeric_data = data.select_dtypes(include=['float64', 'int64'])
statistics = pd.DataFrame({
    'Mean': numeric_data.mean(),
    'Median': numeric_data.median(),
    'Standard Deviation': numeric_data.std(),
    'Skewness': numeric_data.skew()
})
print(statistics)
```

```python
# 1. Bar Chart for Cuisine_Type Distribution
def plot_cuisine_distribution(data):
    """Creates a bar chart of the distribution of restaurants by Cuisine_Type."""
    plt.figure(figsize=(8, 5))
    data['Cuisine_Type'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title('Distribution of Restaurants by Cuisine Type')
    plt.xlabel('Cuisine Type')
    plt.ylabel('Number of Restaurants')
    plt.xticks(rotation=45)
    plt.show()

plot_cuisine_distribution(data)
```

```python
# 2. Scatter Plot for Marketing Spend vs Monthly Revenue
def plot_marketing_vs_revenue(data):
    """Creates a scatter plot of Marketing Spend vs Monthly Revenue."""
    plt.figure(figsize=(8, 5))
    sns.scatterplot(data=data, x='Marketing_Spend', y='Monthly_Revenue', hue='Cuisine_Type', palette='viridis')
    plt.title('Marketing Spend vs Monthly Revenue')
    plt.xlabel('Marketing Spend')
    plt.ylabel('Monthly Revenue')
    plt.legend(title='Cuisine Type')
    plt.show()

plot_marketing_vs_revenue(data)
```

```python
# 3. Heatmap of Correlation Matrix for Numerical Columns
def plot_correlation_heatmap(data):
    """Creates a heatmap of the correlation matrix for numerical features."""
    plt.figure(figsize=(10, 8))
    correlation_matrix = data.corr(numeric_only=True)
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f")
    plt.title('Correlation Matrix Heatmap')
    plt.show()

plot_correlation_heatmap(data)
```
